```python
import pytesseract
import cv2
from PIL import Image
# Read the image
image = cv2.imread('images.jfif')

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Set the Tamil language as the language to be recognized
custom_config = r'--oem 3 --psm 6 tessdata-dir "tam" lang tamil'
text = pytesseract.image_to_string(gray, config=custom_config)

print(text)
```

```python
# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Set the Tesseract OCR engine language to Tamil
config = ('--oem 1 --psm 3 lang=tam')

# Extract the text from the image
text = pytesseract.image_to_string(gray, config=config)

# Print the text
print(text)
```

```python
def process_image(iamge_name, lang_code):
    return pytesseract.image_to_string(Image.open(iamge_name), lang=lang_code)

def print_data(data):
    print(data)

def output_file(filename, data):
    file = open(filename, "w+")
    file.write(data)
    file.close()

def main():
    data_eng = process_image("new3.jpg", "tam")
    #data_ben = process_image("test_ben.png", "ben")
    print_data(data_eng)
    output_file("eng.txt", data_eng)

if  __name__ == '__main__':
	main()
```
